```python

data['diagnosis']=data['diagnosis'].replace('M',0)
data['diagnosis']=data['diagnosis'].replace('B',1)
```

```python
data=data.drop(columns='id')
```

Splitting the data into training data & Testing data

```python
X = data.drop(columns='diagnosis', axis=1)
Y = data['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
```

logistic regression

```python
X_train=st.fit_transform(X_train)
X_test=st.fit_transform(X_test)

```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="mean")

imputer.fit(X_train)

X_train = imputer.transform(X_train)
```

```python
model=LogisticRegression()
```

Accuracy Score

```python
model.fit(X_train, y_train)
```

```python
x_predection=model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
training_data_accuracy=accuracy_score(y_train,x_predection)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="mean")

imputer.fit(X_test)

X_test = imputer.transform(X_test)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(y_test, X_test_prediction)
```

```python
print('accuray on test=',test_data_accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,X_test_prediction)
sns.heatmap(cm, annot=True, xticklabels=['malignant', 'benign'], yticklabels=['malignant', 'benign'])
```
